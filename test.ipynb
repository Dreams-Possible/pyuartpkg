```python
import struct
import time
import pyuart_v5 as uart
```

```python
# 检查串口
ports = uart.list_uart()
# 绑定默认串口
if ports.list:
    defport = ports.list[0].device
else:
    raise RuntimeError("no uart device")
```

```python
# 创建串口1实例并初始化
uartins1 = uart.UartIns()
uartins1.init(ports.list[0].device, 115200)

# 创建串口1通信实例并绑定
comuins1 = uart.UartComuIns()
comuins1.init(uartins1)

time.sleep(0.1)  # 等待串口稳定
```

```python
# 创建串口2实例并初始化
uartins2 = uart.UartIns()
uartins2.init(ports.list[1].device, 115200)

# 创建串口2通信实例并绑定
comuins2 = uart.UartComuIns()
comuins2.init(uartins2)

time.sleep(0.1)  # 等待串口稳定
```

```python

# -----------------------------
# 1️⃣ 发送 float 类型数据（模拟一次传感器 4 个 float）
float_data = [1.23, 4.56, 7.89, 0.12]
print("Send float bytes:", float_data)
data_bytes = struct.pack("<4f", *float_data)
comuins1.send(data_bytes)
time.sleep(0.1)

received = comuins2.recv()
if received:
    print("Received float bytes:", received)
    received_floats = struct.unpack("<4f", received)
    print("Decoded floats:", received_floats)
else:
    print("No float data received")
```
